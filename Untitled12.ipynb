```python
#correlation matrix
df=pd.DataFrame(data)
```

```python
data.isnull().sum()
```

#there is no null values in the data set

```python
#removing outliers
from scipy import stats
a=stats.zscore(data.Age)
data.Age[(a<3)]
sns.boxplot(data.Age)
```

```python
p=stats.zscore(data.Pclass)
data.Pclass[(p<3)]
data.Pclass.head()
sns.boxplot(data.Pclass)
```
